```python
import pandas as pd

df = pd.read_csv('multiclass.csv')
df.describe()
```

```python
from sklearn.model_selection import train_test_split

target = "class"
X = df.drop(target, axis=1)
y = df[target]
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X = scaler.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

```python
import numpy as np

class KNN:
    def __init__(self, k):
        self.k = k

    def predict(self, X_train, y_train, X_test):
      if hasattr(y_train, "iloc"):
          y_train = y_train.reset_index(drop=True)
      y_pred = []
      for sample in X_test:
          dist = np.sqrt(np.sum((X_train - sample) ** 2, axis=1))
          k_indices = np.argsort(dist)[:self.k]
          k_nearest_labels = [y_train.iloc[i] if hasattr(y_train, "iloc") else y_train[i] for i in k_indices]
          classes = {}
          for label in k_nearest_labels:
              classes[label] = classes.get(label, 0) + 1
          y_pred.append(max(classes, key=classes.get))
      return np.array(y_pred)

    def score(self, X_train, y_train, X_test, y_test):
        predictions = self.predict(X_train, y_train, X_test)
        return np.mean(predictions == y_test)
```

```python
all_results = []
for k in range(1, 10, 2):
    print("KNN with k =", k)
    knn = KNN(k=k)
    accuracy = knn.score(X_train, y_train, X_test, y_test)
    results = {"k": k, "accuracy": accuracy}
    all_results.append(results)
    print(results)

best_result = max(all_results, key=lambda x: x["accuracy"])
print("\nFinal Results:", all_results)
print("Best k =", best_result["k"], "with accuracy =", best_result["accuracy"])
```

Using Sklearn

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

k_range = range(1, 21)
accuracies = []

for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    accuracies.append(acc)
```

```python
optimal_k = k_range[accuracies.index(max(accuracies))]
print(f"Optimal K value: {optimal_k}")
print(f"Best Accuracy: {max(accuracies):.2f}")
```
